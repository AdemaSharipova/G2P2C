```python
from debug_visualiser import get_debug_info
from matplotlib import pyplot as plt
import matplotlib.gridspec
import numpy as np
import pandas as pd
import json
%matplotlib inline

MAIN_PATH = '../results/check3/'
training_workers = 10
training_seeds = [x for x in range(0, training_workers)]
testing_workers = 5
testing_seeds = [500+x for x in range(0, testing_workers)]


def get_avg(path, seeds, filename, column):
    path = path + filename
    cur_length, full_arr, refined = [], [], []
    FILES = [ path + str(seed)+'.csv' for seed in seeds]
    for file in FILES:
        reward_summary = pd.read_csv(file)
        cur_length.append(reward_summary.shape[0])
        full_arr.append(reward_summary[column])
    for x in full_arr:
        refined.append(x[0:min(cur_length)])
    data = pd.concat(refined, axis=1)
    data['mean'] = data.mean(axis=1)
    data['std_dev'] = data.std(axis=1)
    return data

#### print the training episode rewards
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)
data = get_avg(MAIN_PATH, training_seeds, '/reward_summary_', 'reward')
mstd = data['std_dev']
ma = data['mean'] 
ax.plot(ma, '-b', label="PPO Training Workers")
ax.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color="b", alpha=0.2)

ax.grid()
ax.set_title('PPO Training - Episodic Rewards')
ax.set_xlabel('Updates')
ax.set_ylabel('Average Reward')
ax.legend(loc="upper right")
plt.show()

#### print the testing episode rewards
fig = plt.figure(figsize=(16, 8))
ax = fig.add_subplot(111)
data = get_avg(MAIN_PATH, testing_seeds, '/testing_episode_summary_', 'reward')
mstd = data['std_dev']
ma = data['mean'] 
ax.plot(ma, '-b', label="PPO Testing Workers")
ax.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color="b", alpha=0.2)

ax.grid()
ax.set_title('PPO Testing - Episodic Rewards')
ax.set_xlabel('Episode')
ax.set_ylabel('Reward')
ax.legend(loc="upper right")
plt.show()


```

```python
#### action history.... 

def get_concat(path, seeds, filename, column):
    path = path + filename
    full_arr, refined = [], []
    FILES = [ path + str(seed)+'.csv' for seed in seeds]
    for file in FILES:
        d = pd.read_csv(file)
        full_arr.append(d[column])
    data = pd.concat(full_arr, axis=0)
    return data

fig = plt.figure(figsize=(16, 14))
ax = fig.add_subplot(411)
ax1 = fig.add_subplot(412)
ax2 = fig.add_subplot(413)
ax3 = fig.add_subplot(414)

data1 = get_concat(MAIN_PATH, training_seeds, 'action_summary_', 'action')
ax.hist(data1, bins = 40)
ax.set_title('Agent Action')

data2 = get_concat(MAIN_PATH, training_seeds, 'action_summary_', 'mean')
ax1.hist(data2, bins = 40)
ax1.set_title('Agent Action Mean')

data3 = get_concat(MAIN_PATH, training_seeds, 'action_summary_', 'pump_action')
ax2.hist(data3, bins = 20)
ax2.set_title('Final Pump Action')

data4 = get_concat(MAIN_PATH, testing_seeds, 'test_action_summary_', 'pump_action')
ax3.hist(data4, bins = 20)
ax3.set_title('Final Pump Action - Testing Simulations')

plt.show()



# print the testing episode rewards
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111)
data = get_avg(MAIN_PATH, testing_seeds, '/testing_episode_summary_', 'ri')
mstd = data['std_dev']
ma = data['mean'] 
ax.plot(ma, '-b', label="RI")
ax.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color="b", alpha=0.2)

data = get_avg(MAIN_PATH, testing_seeds, '/testing_episode_summary_', 'normo')
mstd = data['std_dev']
ma = data['mean'] 
ax.plot(ma, '-g', label="Time in Target Glucose Range")
ax.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color="g", alpha=0.2)

ax.grid()
ax.set_title('PPO Testing - Episodic Rewards')
ax.set_xlabel('Episode')
ax.set_ylabel('Reward')
ax.legend(loc="upper right")
plt.show()
```
