```python
# the episode wise information.

def get_episodic_debug_info(path, seeds):
    cur_length = []
    data_frames = []
    FILES = [ path + str(seed)+'.csv' for seed in seeds]
    for file in FILES:
        action_summary = pd.read_csv(file)
        cur_length.append(action_summary.shape[0])
        data_frames.append(action_summary)
    print(data_frames)
        
path_a2c = '../results/exp_22March_2/training_summary_'
get_episodic_debug_info(path_a2c, seeds)
        
```
