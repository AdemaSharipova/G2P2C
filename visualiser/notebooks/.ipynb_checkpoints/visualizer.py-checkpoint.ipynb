```python
from matplotlib import pyplot as plt
import pandas as pd
import json
%matplotlib inline

TARGET_FOLDER = 'testing_basal_only'
PATH = '../results/testing_basal_only/args.json'
FILE = '../results/testing_basal_only/training_summary.csv'


with open(PATH) as json_file:
    data = json.load(json_file)
print(data)

print(data['run_mode'])

training_summary = pd.read_csv(FILE)

# print summary of the training stats.
```

```python
from matplotlib import pyplot as plt
import pandas as pd
import json
%matplotlib inline

FILE = '../results/testing_basal_only/avg_reward_summary.csv'
reward_summary = pd.read_csv(FILE)

fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(111)

reward_hist = reward_summary['avg_reward_step']

ax.plot(reward_hist, '-b')
ax.set_title('Average Rewards vs Steps')
ax.set_xlabel('Steps')
ax.set_ylabel('Average Reward')
plt.show()

```

```python
fig = plt.figure(figsize=(16, 10))
ax = fig.add_subplot(211)
ax1 = fig.add_subplot(212)
ax.plot(training_summary['Average Reward'], '-b')
mstd = training_summary['t_rew_std']
ma = training_summary['t_rew_mean']
ax.plot(ma, '-g')
ax.fill_between(mstd.index, ma - 2 * mstd, ma + 2 * mstd, color="g", alpha=0.2)
ax.set_title('Average Rewards vs Episodes')
ax.set_xlabel('Episodes')
ax.set_ylabel('Average Reward')
ax1.plot(training_summary['RI'], '-b')
mstd_ri = training_summary['t_ri_std']
ma_ri = training_summary['t_ri_mean']
ax1.plot(ma_ri, '-g')
ax1.fill_between(mstd_ri.index, ma_ri - 2 * mstd_ri, ma_ri + 2 * mstd_ri, color="g", alpha=0.2)
ax1.set_title('Risk Index vs Episodes')
ax1.set_xlabel('Episodes')
ax1.set_ylabel('Risk Index')
plt.show()

cur_interactions = training_summary['Current Interactions'].iloc[-1]
cur_net_updates = training_summary['Network Updates'].iloc[-1]
cur_episode = training_summary['Episode'].iloc[-1]
print('Training has run for {} episodes, {}/{} steps, Network Updates {}'.format(cur_episode, cur_interactions, data['interactions'],cur_net_updates))
```

```python

import numpy as np
import ast

DISPLAY_EPISODE = -1

glucose_hist = training_summary['glucose_hist'].iloc[DISPLAY_EPISODE]
meal_hist = training_summary['meal_hist'].iloc[DISPLAY_EPISODE]
insulin_hist = training_summary['insulin_hist'].iloc[DISPLAY_EPISODE]

glucose_hist = ast.literal_eval(glucose_hist)
meal_hist = ast.literal_eval(meal_hist)
insulin_hist = ast.literal_eval(insulin_hist)

# todo: restrict to 24 hours. 

fig = plt.figure(figsize=(16, 10))
fig.suptitle(' Summary results for 24hours, Episode:')
ax = fig.add_subplot(311)  # The big subplot
ax1 = fig.add_subplot(312)
ax2 = fig.add_subplot(313)
ax.plot(glucose_hist)
ax.axhline(y=40, color='r', linestyle='--')
ax.axhline(y=70, color='g', linestyle='--')
ax.axhline(y=180, color='g', linestyle='--')
ax1.plot(meal_hist)
ax2.plot(insulin_hist)
ax.set_title('Glucose')
ax1.set_title('Meals')
ax2.set_title('Insulin')
ax.set(ylabel='Glucose (mg/DL)')
ax1.set(ylabel='Carbs (grams)')
ax2.set(ylabel='Insulin (units)', xlabel='Time')
# ax2.axhline(y=STD_BASAL, color='r', linestyle='--')
plt.show()
```
